### Análisis del Fenómeno: La Súper Estabilidad del Random Forest

Si miras el reporte `Estabilidad_Variables_RF_MORTALIDAD.csv` que me enviaste, **absolutamente todas las variables (las 110) tienen una "Tasa_Seleccion_Porcentaje" del 100.0%**. Es decir, en las 100 repeticiones del Bootstrapping, el Random Forest siempre encontró alguna utilidad para todas y cada una de las variables clínicas.

**¿Por qué ocurrió esto?**
Esto no es un error de tu código, es la firma de identidad del algoritmo *Random Forest*.
1.  **Muestreo de Variables (*Feature Bagging*):** Random Forest no evalúa todas las variables al mismo tiempo. En cada división de cada uno de sus 200 árboles, selecciona un subconjunto aleatorio de variables (generalmente la raíz cuadrada del total, unas 10 variables por nodo). Esto obliga al modelo a "darle la oportunidad" a variables débiles de ser utilizadas si por azar no compiten contra variables fuertes como `NUM_PROCEDIMIENTOS` en esa división específica.
2.  **Dataset Masivo:** Tienes casi 800.000 filas. Con tanta información, incluso la especialidad médica más rara (ej. Odontología) tiene suficientes casos para que un árbol profundo justifique usarla en alguna ramificación remota.

**¿Qué significa esto para tu Tesis?**
Significa que el filtro de "frecuencia de selección" (quedarnos con las que salen muchas veces) **no aplica** porque todas salen siempre. Por lo tanto, debemos pasar al "Plan B" metodológico: **filtrar por magnitud de importancia.**

````python
import pandas as pd # Importa librería para análisis de datos
import numpy as np # Importa librería para cálculos numéricos
import os # Importa módulo para interactuar con el sistema de archivos
import gc # Importa recolector de basura para liberar RAM
import time # Importa módulo para medir tiempos de ejecución
from sklearn.ensemble import RandomForestClassifier # Importa el algoritmo campeón para Mortalidad
from sklearn.metrics import f1_score, average_precision_score # Importa métricas de evaluación

def seleccion_caracteristicas_bootstrapping_rf_mortalidad():
    """
    - Descripción: Aplica el método de Stability Selection mediante Bootstrapping (100 repeticiones)
                   sobre el modelo campeón de Random Forest para la variable MORTALIDAD.
                   Evalúa las variables en base a su frecuencia de selección e importancia promedio.
    - Salida: CSV con el reporte de estabilidad de características y métricas OOB.
    """
    target_name = 'MORTALIDAD'
    n_repeticiones = 100 # Paso 1: Definir la cantidad de repeticiones de la muestra
    
    dir_datos = "../../Datos/Datasets Finales"
    dir_resultados = "../../Resultados/Resultados (etapa 4)/Seleccion_Bootstrapping"
    os.makedirs(dir_resultados, exist_ok=True) # Crea la carpeta para esta nueva etapa si no existe

    cols_excluir = ['CONSUMO_RECURSOS', 'SEVERIDAD', 'MORTALIDAD', 'CATEGORIA_CANCER'] 

    print("="*60)
    print(f"INICIANDO BOOTSTRAPPING (100 ITERS) PARA: {target_name} (RANDOM FOREST)")
    print("="*60)

    # Cargar datos de entrenamiento
    print("Cargando datasets...")
    df_onco_train = pd.read_csv(os.path.join(dir_datos, "dataset_entrenamiento_onco_2019_2023.csv"), low_memory=False)
    df_control_train = pd.read_csv(os.path.join(dir_datos, "dataset_entrenamiento_control_2019_2023.csv"), low_memory=False)

    # Paso 2.2: Balanceo de la muestra maestra (Submuestreo del control)
    print("Balanceando la cohorte basal...")
    n_onco = len(df_onco_train)
    df_control_sample = df_control_train.sample(n=n_onco, random_state=42)
    df_train_maestro = pd.concat([df_onco_train, df_control_sample], ignore_index=True)

    del df_onco_train, df_control_train, df_control_sample
    gc.collect() # Libera RAM pesada antes del bucle

    features = [col for col in df_train_maestro.columns if col not in cols_excluir]
    
    # Diccionarios y listas para almacenar la memoria de las 100 iteraciones (Paso 2.6)
    memoria_importancias = {feature: [] for feature in features}
    frecuencia_seleccion = {feature: 0 for feature in features}
    registro_metricas_oob = []

    print(f"Iniciando bucle de {n_repeticiones} iteraciones. ¡Esto tomará tiempo!")
    tiempo_inicio_total = time.time()

    indices_totales = df_train_maestro.index

    # BUCLE PRINCIPAL DE BOOTSTRAPPING
    for i in range(n_repeticiones):
        inicio_iter = time.time()
        
        # Paso 2.1: Tomar una muestra aleatoria con reemplazo del dataset (Misma longitud, permite repetidos)
        df_boot = df_train_maestro.sample(frac=1.0, replace=True, random_state=i)
        
        # Identificar los datos OOB (Out-Of-Bag) - Los que no entraron en esta muestra
        indices_oob = indices_totales.difference(df_boot.index)
        df_oob = df_train_maestro.loc[indices_oob]
        
        # Separar X e Y para la bolsa de entrenamiento
        X_boot = df_boot[features]
        y_boot = df_boot[target_name]
        
        # Separar X e Y para la bolsa de evaluación (OOB)
        X_oob = df_oob[features]
        y_oob = df_oob[target_name]
        
        # Paso 2.3: Entrenar el modelo con los hiperparámetros ganadores
        modelo_rf = RandomForestClassifier(
            n_estimators=200, 
            max_depth=None, 
            min_samples_split=10, 
            class_weight='balanced',
            n_jobs=-1, # Usa todos los hilos del procesador
            random_state=42 # Fija la semilla matemática interna
        )
        modelo_rf.fit(X_boot, y_boot)
        
        # Paso 2.4: Ver qué variables selecciona y qué tan importantes son
        importancias = modelo_rf.feature_importances_
        
        for idx, col in enumerate(features):
            imp_val = importancias[idx]
            memoria_importancias[col].append(imp_val) # Guarda el valor numérico
            if imp_val > 0: # Si el árbol la usó al menos una vez para dividir nodos
                frecuencia_seleccion[col] += 1
                
        # Paso 2.5: Evaluar el modelo en los datos que no fueron usados (OOB)
        y_pred_oob = modelo_rf.predict(X_oob)
        y_prob_oob = modelo_rf.predict_proba(X_oob)[:, 1] # Toma la probabilidad de la clase 1 (Fallecimiento)
        
        f1_macro_oob = f1_score(y_oob, y_pred_oob, average='macro')
        auprc_oob = average_precision_score(y_oob, y_prob_oob)
        
        registro_metricas_oob.append({'Iteracion': i+1, 'F1_Macro_OOB': f1_macro_oob, 'AUPRC_OOB': auprc_oob})
        
        fin_iter = time.time()
        print(f"   -> Iteración {i+1}/{n_repeticiones} completada en {round(fin_iter - inicio_iter, 1)} seg. | OOB F1-Macro: {f1_macro_oob:.4f}")

    tiempo_fin_total = time.time()
    print(f"\nÉXITO: Bootstrapping completado en {round((tiempo_fin_total - tiempo_inicio_total)/60, 2)} minutos.")

    # =========================================================================
    # FASE 3: CONSOLIDACIÓN Y ALMACENAMIENTO DE RESULTADOS
    # =========================================================================
    
    # 3.1 y 3.2: Contar apariciones y promediar importancias
    resultados_finales = []
    for col in features:
        tasa_aparicion = (frecuencia_seleccion[col] / n_repeticiones) * 100 # Porcentaje de veces que sobrevivió
        imp_promedio = np.mean(memoria_importancias[col]) # Importancia media en las 100 vueltas
        imp_std = np.std(memoria_importancias[col]) # Desviación estándar de su importancia
        
        resultados_finales.append({
            'Variable': col,
            'Tasa_Seleccion_Porcentaje': tasa_aparicion,
            'Importancia_Promedio': imp_promedio,
            'Varianza_Importancia': imp_std
        })
        
    df_resultados = pd.DataFrame(resultados_finales)
    # 3.3: Ordenar para quedarse visualmente con las más estables en la cima
    df_resultados = df_resultados.sort_values(by=['Tasa_Seleccion_Porcentaje', 'Importancia_Promedio'], ascending=[False, False])
    
    # Guardar reporte de variables
    ruta_vars = os.path.join(dir_resultados, f"Estabilidad_Variables_RF_{target_name}.csv")
    df_resultados.to_csv(ruta_vars, index=False)
    
    # Guardar reporte de métricas OOB
    df_metricas = pd.DataFrame(registro_metricas_oob)
    ruta_metricas = os.path.join(dir_resultados, f"Metricas_OOB_RF_{target_name}.csv")
    df_metricas.to_csv(ruta_metricas, index=False)
    
    print("="*60)
    print(f"- Reporte de Estabilidad guardado en: {ruta_vars}")
    print(f"- Métricas OOB guardadas en: {ruta_metricas}")
    print(f"Métrica OOB Promedio Final -> F1-Macro: {df_metricas['F1_Macro_OOB'].mean():.4f} | AUPRC: {df_metricas['AUPRC_OOB'].mean():.4f}")
    print("="*60)

# Para ejecutar:
# seleccion_caracteristicas_bootstrapping_rf_mortalidad()
````

### SEVERIDAD Y CONSUMO DE RECURSOS

````python
import pandas as pd
import numpy as np
import os
import gc
import time
import xgboost as xgb
from sklearn.metrics import f1_score, average_precision_score
from sklearn.preprocessing import label_binarize

def seleccion_caracteristicas_bootstrapping_xgb_multiclase(target_name):
    """
    - Descripción: Aplica Stability Selection (Bootstrapping, 100 repeticiones)
                   sobre el modelo campeón XGBoost para variables MULTICLASE.
    - Entrada: target_name ('SEVERIDAD' o 'CONSUMO_RECURSOS').
    - Salida: CSV con estabilidad de variables y métricas OOB.
    """
    n_repeticiones = 100
    dir_datos = "../../Datos/Datasets Finales"
    dir_resultados = "../../Resultados/Resultados (etapa 4)/Seleccion_Bootstrapping"
    os.makedirs(dir_resultados, exist_ok=True)

    cols_excluir = ['CONSUMO_RECURSOS', 'SEVERIDAD', 'MORTALIDAD', 'CATEGORIA_CANCER', 'DIAS_ESTADIA']

    print("="*60)
    print(f"INICIANDO BOOTSTRAPPING MULTICLASE PARA: {target_name} (XGBOOST)")
    print("="*60)

    print("Cargando datasets...")
    df_onco_train = pd.read_csv(os.path.join(dir_datos, "dataset_entrenamiento_onco_2019_2023.csv"), low_memory=False)
    df_control_train = pd.read_csv(os.path.join(dir_datos, "dataset_entrenamiento_control_2019_2023.csv"), low_memory=False)

    print("Balanceando la cohorte basal...")
    n_onco = len(df_onco_train)
    df_control_sample = df_control_train.sample(n=n_onco, random_state=42)
    df_train_maestro = pd.concat([df_onco_train, df_control_sample], ignore_index=True)

    del df_onco_train, df_control_train, df_control_sample
    gc.collect()

    features = [col for col in df_train_maestro.columns if col not in cols_excluir]
    
    # Identificar clases para el binarizador OOB
    clases_unicas = np.unique(df_train_maestro[target_name])

    memoria_importancias = {feature: [] for feature in features}
    frecuencia_seleccion = {feature: 0 for feature in features}
    registro_metricas_oob = []

    print(f"Iniciando bucle de {n_repeticiones} iteraciones...")
    tiempo_inicio_total = time.time()
    indices_totales = df_train_maestro.index

    for i in range(n_repeticiones):
        inicio_iter = time.time()
        
        # 1. Bootstrapping
        df_boot = df_train_maestro.sample(frac=1.0, replace=True, random_state=i)
        
        # 2. OOB
        indices_oob = indices_totales.difference(df_boot.index)
        df_oob = df_train_maestro.loc[indices_oob]
        
        X_boot = df_boot[features]
        y_boot = df_boot[target_name]
        
        X_oob = df_oob[features]
        y_oob = df_oob[target_name]
        
        # 3. Entrenar XGBoost Multiclase con hiperparámetros óptimos
        modelo_xgb = xgb.XGBClassifier(
            learning_rate=0.3, # Hiperparámetro campeón
            max_depth=10, # Hiperparámetro campeón
            tree_method='hist',
            n_jobs=-1,
            random_state=42
        )
        modelo_xgb.fit(X_boot, y_boot)
        
        # 4. Extraer importancias
        importancias = modelo_xgb.feature_importances_
        
        for idx, col in enumerate(features):
            imp_val = importancias[idx]
            memoria_importancias[col].append(imp_val)
            if imp_val > 0:
                frecuencia_seleccion[col] += 1
                
        # 5. Evaluar OOB (Lógica Multiclase OvR Ponderada)
        y_pred_oob = modelo_xgb.predict(X_oob)
        y_prob_oob = modelo_xgb.predict_proba(X_oob)
        
        f1_macro_oob = f1_score(y_oob, y_pred_oob, average='macro')
        
        # Binarizar para calcular AUPRC multiclase ponderado
        y_oob_bin = label_binarize(y_oob, classes=clases_unicas)
        auprc_oob = average_precision_score(y_oob_bin, y_prob_oob, average='weighted')
        
        registro_metricas_oob.append({'Iteracion': i+1, 'F1_Macro_OOB': f1_macro_oob, 'AUPRC_OOB_Weighted': auprc_oob})
        
        fin_iter = time.time()
        print(f"   -> Iteración {i+1}/{n_repeticiones} completada en {round(fin_iter - inicio_iter, 1)} seg. | OOB F1: {f1_macro_oob:.4f}")

    tiempo_fin_total = time.time()
    print(f"\n- Bootstrapping completado en {round((tiempo_fin_total - tiempo_inicio_total)/60, 2)} minutos.")

    # Consolidación final
    resultados_finales = []
    for col in features:
        tasa_aparicion = (frecuencia_seleccion[col] / n_repeticiones) * 100
        imp_promedio = np.mean(memoria_importancias[col])
        imp_std = np.std(memoria_importancias[col])
        
        resultados_finales.append({
            'Variable': col,
            'Tasa_Seleccion_Porcentaje': tasa_aparicion,
            'Importancia_Promedio': imp_promedio,
            'Varianza_Importancia': imp_std
        })
        
    df_resultados = pd.DataFrame(resultados_finales)
    df_resultados = df_resultados.sort_values(by=['Tasa_Seleccion_Porcentaje', 'Importancia_Promedio'], ascending=[False, False])
    
    ruta_vars = os.path.join(dir_resultados, f"Estabilidad_Variables_XGB_{target_name}.csv")
    df_resultados.to_csv(ruta_vars, index=False)
    
    df_metricas = pd.DataFrame(registro_metricas_oob)
    ruta_metricas = os.path.join(dir_resultados, f"Metricas_OOB_XGB_{target_name}.csv")
    df_metricas.to_csv(ruta_metricas, index=False)
    
    print("="*60)
    print(f"- Reporte de Estabilidad guardado en: {ruta_vars}")
    print("="*60)
````
